# Exponential kernel eigen-decomposition

```python
from kl_decomposition import exp_kernel
alpha = 2.0
vals_c, freqs_c, vals_s, freqs_s = exp_kernel.exp_kernel_eigen(alpha, L=1.0, n_cos=3, n_sin=3)
print('cos eigenvalues:', vals_c)
print('sin eigenvalues:', vals_s)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,1,200)
plt.figure()
for w in freqs_c:
    plt.plot(x, np.sqrt(2)*np.cos(w*x), label=f'cos w={w:.2f}')
for w in freqs_s:
    plt.plot(x, np.sqrt(2)*np.sin(w*x), '--', label=f'sin w={w:.2f}')
plt.legend()
plt.show()
```
